```python
import pandas as pd
import re
import os, glob
Elements = []
filePath = ["business","entertainment","politics","sport","tech"]
for i in filePath:
    path = i
    for bmp_file in glob.glob(os.path.join(path, '*')):
        my_file = open(bmp_file)
        print(bmp_file)
        my_string = my_file.read()
        reg1 = ["[A-Z][a-z]+\s+[A-Z][a-z]+\s+[A-Z][a-z]+\s+[A-Z][a-z]+","[A-Z][a-z]+\s+[A-Z][a-z]+\s+[a-z]+\s+[A-Z][a-z]+","\?([A-Z][a-z]+\s+[A-Z][a-z]+\s+[A-Z][a-z]+)","\?([A-Z][a-z]+\s+[A-Z][a-z]+)"]
        
    
        for r in reg1:
            regex = re.compile(r, re.UNICODE)
            number = regex.findall(my_string)
            Elements.append(number)
            print(number)
            
        
```
