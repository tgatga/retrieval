```python
names=[("Aslan","Male"),
       ("Daulet","Male"),
       ("Ernar","Male"),
       ("Meirambek","Male"),
       ("Berik","Male"),
       ("Kasym","Male"),
       ("Shyngys","Male"),
       ("Erzhan","Male"),
       ("Daulet","Male"),
       ("Assel","Female"),
       ("Aruzhan","Female"),
       ("Aigerim","Female"),
       ("Anar","Female"),
       ("Dina","Female"),
       ("Zhanara","Female"),
       ("Zarina","Female"),
       ("Zhannur","Female"),
       ("Balausa","Female"),
       ("Nazerke","Female"),
       ("Alma","Female")]
print(len(names))
def gender_features(word):
   return {'last_letter': word[-1]}
 
gender_features('Assel')
```

```python
featuresets = [(gender_features(n), g) for (n, g) in names]
featuresets
```

```python
train_set, test_set = featuresets[1:], featuresets[:20]
```

```python
from nltk import NaiveBayesClassifier
nb_classifier = NaiveBayesClassifier.train(train_set)
print(nb_classifier.classify(gender_features('Dina')))
from nltk import classify
classify.accuracy(nb_classifier, test_set)
```

```python
from nltk import MaxentClassifier
me_classifier = MaxentClassifier.train(train_set)
print(me_classifier.classify(gender_features('Laura')))
classify.accuracy(me_classifier, test_set)
```

```python
def gender_features2(name):
    features = {}
    features["firstletter"] = name[0].lower()
    features["lastletter"] = name[-1].lower()
    for letter in 'abcdefghijklmnopqrstuvwxyz':
            features["count(%s)" % letter] = name.lower().count(letter)
            features["has(%s)" % letter] = (letter in name.lower())
    return features
featuresets = [(gender_features2(n), g) for (n, g) in names]
train_set, test_set = featuresets[10:], featuresets[:10]
nb2_classifier = NaiveBayesClassifier.train(train_set)
print(classify.accuracy(nb2_classifier, test_set))
me2_classifier = MaxentClassifier.train(train_set)
classify.accuracy(me2_classifier, test_set)
    
```

```python
def gender_features3(name):
    features = {}
    features["fl"] = name[0].lower()
    features["ll"] = name[-1].lower()
    features["fw"] = name[:2].lower()
    features["lw"] = name[-2:].lower()
    return features
featuresets = [(gender_features3(n), g) for (n, g) in names]
train_set, test_set = featuresets[10:], featuresets[:10]
nb3_classifier = NaiveBayesClassifier.train(train_set)
print(classify.accuracy(nb3_classifier, test_set))
me3_classifier = MaxentClassifier.train(train_set)
classify.accuracy(me3_classifier, test_set)
```
